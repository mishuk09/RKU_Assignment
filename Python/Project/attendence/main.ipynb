```python
import mysql.connector
from tkinter import *

# Establish a connection to the MySQL database
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="rku_attendence_admin"
)

# Create a cursor to execute SQL statements
mycursor = mydb.cursor()

# Create the CE students table if it doesn't exist
mycursor.execute("CREATE TABLE IF NOT EXISTS ce_students (name VARCHAR(255), roll_number INT)")

# Create the IT students table if it doesn't exist
mycursor.execute("CREATE TABLE IF NOT EXISTS it_students (name VARCHAR(255), roll_number INT)")

# Define the function that will be called when the "Register" button is clicked
def register_student():
    # Get the values of the entry widgets
    name = name_entry.get()
    roll_number = roll_number_entry.get()
    department = department_variable.get()

    # Insert the student's details into the appropriate table
    if department == "CE":
        mycursor.execute("INSERT INTO ce_students (name, roll_number) VALUES (%s, %s)", (name, roll_number))
    elif department == "IT":
        mycursor.execute("INSERT INTO it_students (name, roll_number) VALUES (%s, %s)", (name, roll_number))

    # Commit the changes to the database
    mydb.commit()

    # Clear the entry widgets
    name_entry.delete(0, END)
    roll_number_entry.delete(0, END)

# Create the main window and widgets
window = Tk()
window.title("Student Registration Form")

name_label = Label(window, text="Name")
name_label.grid(row=0, column=0)

name_entry = Entry(window)
name_entry.grid(row=0, column=1)

roll_number_label = Label(window, text="Roll Number")
roll_number_label.grid(row=1, column=0)

roll_number_entry = Entry(window)
roll_number_entry.grid(row=1, column=1)

department_label = Label(window, text="Department")
department_label.grid(row=2, column=0)

department_variable = StringVar(window)
department_variable.set("CE")

department_dropdown = OptionMenu(window, department_variable, "CE", "IT")
department_dropdown.grid(row=2, column=1)

register_button = Button(window, text="Register", command=register_student)
register_button.grid(row=3, column=0)

window.mainloop()

# Close the database connection
mydb.close()
```
