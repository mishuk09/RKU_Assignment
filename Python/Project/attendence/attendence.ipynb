```python
import tkinter as tk
import mysql.connector as mysql
from tkinter import ttk

# Connect to MySQL database
db = mysql.connect(
    host="localhost",
    user="root",
    password="",
    database="rku_attendence_admin"
)

# Create a Tkinter window
window = tk.Tk()
window.title("RKU Attendance System")

# Create a label for the window
welcome_label = tk.Label(window, text="Welcome to RKU Server", font=("Arial", 18))
welcome_label.pack(pady=10)

# Create a label for the 'Faculty' input
faculty_label = tk.Label(window, text="Faculty", font=("Arial", 12))
faculty_label.pack(pady=5)

# Create an entry box for the 'Faculty' input
faculty_entry = tk.Entry(window, font=("Arial", 12))
faculty_entry.pack(pady=5)

# Create a label for the 'Date' input
date_label = tk.Label(window, text="Date", font=("Arial", 12))
date_label.pack(pady=5)

# Create an entry box for the 'Date' input
date_entry = tk.Entry(window, font=("Arial", 12))
date_entry.pack(pady=5)

# Create a label for the dropdown
division_label = tk.Label(window, text="Division", font=("Arial", 12))
division_label.pack(pady=5)

# Create a dropdown with two options: IT and CE
division_options = ["IT", "CE"]
division_var = tk.StringVar(window)
division_dropdown = tk.OptionMenu(window, division_var, *division_options)
division_dropdown.pack(pady=5)

# Create a tick mark system for taking attendance
def take_attendance():
    # Get the selected division
    division = division_var.get()

    # Get the students from the database
    cursor = db.cursor()
    cursor.execute(f"SELECT * FROM {division.lower()}_students")
    students = cursor.fetchall()

    # Create a new window for taking attendance
    attendance_window = tk.Toplevel(window)
    attendance_window.title(f"{division} Attendance")

    # Create a label for the window
    attendance_label = tk.Label(attendance_window, text=f"{division} Attendance", font=("Arial", 18))
    attendance_label.pack(pady=10)

    # Create a Treeview widget for displaying the attendance data
    attendance_tree = ttk.Treeview(attendance_window, columns=("name", "attendance"), show="headings")
    attendance_tree.heading("name", text="Name")
    attendance_tree.heading("attendance", text="Attendance")
    attendance_tree.pack()

    # Add the students to the Treeview widget
    for student in students:
        attendance_tree.insert("", "end", values=(student[1], ""))

        # Save the attendance in the database
        def save_attendance():
            attendance = 1 if attendance_var.get() else 0
            cursor.execute(f"INSERT INTO {division.lower()}_attendance (student_id, date, attendance) VALUES ({student[0]}, '{date_entry.get()}', {attendance})")
            db.commit()

        # Create a tick mark system for the student's attendance
        attendance_var = tk.BooleanVar()
        attendance_checkbox = tk.Checkbutton(attendance_window, variable=attendance_var)
        attendance_checkbox.pack(pady=5)

        # Bind the tick mark system to the Treeview widget
        def update_treeview(*args):
            if attendance_var.get():
                attendance_tree.set(student[1], "attendance", "Present")
            else:
                attendance_tree.set(student[1], "attendance", "Absent")
        
        # Create a Treeview widget for displaying the attendance data
        attendance_tree = ttk.Treeview(attendance_window, columns=("attendance",))
        attendance_tree.heading("#0", text="Name")
        attendance_tree.heading("attendance", text="Attendance")
        attendance_tree.pack(pady=10)
        
        # Insert the students into the Treeview widget
        for student in students:
            attendance_tree.insert("", "end", text=student[1], values=("Absent",))
        
            # Bind the tick mark system to the Treeview widget
            attendance_var = tk.BooleanVar()
            attendance_var.trace("w", update_treeview)
            attendance_checkbox = tk.Checkbutton(attendance_window, variable=attendance_var)
            attendance_tree.window_create(attendance_tree.identify_row(student[1]), column=1, window=attendance_checkbox)
    # Create a button to take attendance
attendance_button = tk.Button(window, text="Take Attendance", command=take_attendance)
attendance_button.pack(pady=10)

# Start the Tkinter event loop
window.mainloop()    
```

```python
import tkinter as tk
import mysql.connector

# Create a connection to the database
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="rku_attendence_admin"
)

# Create a cursor object to execute SQL queries
mycursor = mydb.cursor()

# Create a table if it does not already exist
mycursor.execute("CREATE TABLE IF NOT EXISTS attendance (id INT AUTO_INCREMENT PRIMARY KEY, faculty VARCHAR(255), division VARCHAR(255), date VARCHAR(255), student_name VARCHAR(255), attendance_status VARCHAR(255))")

class AttendanceGUI:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("RKU Attendance System")

        # Create the welcome label
        welcome_label = tk.Label(self.root, text="Welcome to RKU Server", font=("Arial", 14))
        welcome_label.grid(row=0, column=0, columnspan=3, pady=10)

        # Create input fields for faculty, division, and date
        faculty_label = tk.Label(self.root, text="Faculty:")
        faculty_label.grid(row=1, column=0, padx=10, pady=10, sticky="w")
        self.faculty_entry = tk.Entry(self.root)
        self.faculty_entry.grid(row=1, column=1, padx=10, pady=10)
 

        date_label = tk.Label(self.root, text="Date:")
        date_label.grid(row=3, column=0, padx=10, pady=10, sticky="w")
        self.date_entry = tk.Entry(self.root)
        self.date_entry.grid(row=3, column=1, padx=10, pady=10)

        division_label = tk.Label(self.root, text="Division:")
        division_label.grid(row=2, column=0, padx=10, pady=10, sticky="w")
        self.division_var = tk.StringVar(value="IT")  # Set the default value to IT
        self.division_dropdown = tk.OptionMenu(self.root, self.division_var, "IT", "CE")  # Create the dropdown menu
        self.division_dropdown.grid(row=2, column=1, padx=10, pady=10)


        # Create the headers for roll, name, and attendance
        roll_label = tk.Label(self.root, text="Roll")
        roll_label.grid(row=4, column=0, padx=10, pady=10)

        name_label = tk.Label(self.root, text="Name")
        name_label.grid(row=4, column=1, padx=10, pady=10)

        attendance_label = tk.Label(self.root, text="Attendance")
        attendance_label.grid(row=4, column=2, padx=10, pady=10)

        # Create the student labels, roll numbers, and tick mark buttons
        self.student_labels = []
        self.roll_labels = []
        self.tick_buttons = []
        for i in range(10):
            roll = i+1
            roll_label = tk.Label(self.root, text=roll)
            roll_label.grid(row=i+5, column=0, padx=10, pady=10)
            self.roll_labels.append(roll_label)

            name_label = tk.Label(self.root, text=f"Student {i+1}")
            name_label.grid(row=i+5, column=1, padx=10, pady=10)
            self.student_labels.append(name_label)

            tick_var = tk.BooleanVar()
            button = tk.Checkbutton(self.root, variable=tick_var)
            button.grid(row=i+5, column=2)
            self.tick_buttons.append(tick_var)

        # Create the submit button
        submit_button = tk.Button(self.root, text="Submit", command=self.submit_attendance)
        submit_button.grid(row=16, column=0, columnspan=3, padx=10, pady=10)

        # Create the show data button
        show_data_button = tk.Button(self.root, text="Show Attendance Data", command=self.show_attendance_data)
        show_data_button.grid(row=18, column=0, columnspan=3, padx=10, pady=10)

    def show_attendance_data(self):
        # Query the attendance data from the database
        mycursor.execute("SELECT * FROM attendance")
        attendance_data = mycursor.fetchall()

        # Create a new window to show the attendance data
        data_window = tk.Toplevel(self.root)
        data_window.title("Attendance Data")
    
        # Create a table to display the attendance data
        table = tk.Frame(data_window)
        table.pack(padx=10, pady=10)
        
        # Create headers for the table
        headers = ["ID", "Faculty", "Division", "Date", "Name", "Attendance Status"]
        for i, header in enumerate(headers):
            label = tk.Label(table, text=header, font=("Arial", 12, "bold"), padx=10, pady=10, borderwidth=1, relief="solid")
            label.grid(row=0, column=i)
    
        # Add the attendance data to the table
        for i, row in enumerate(attendance_data):
            for j, item in enumerate(row):
                label = tk.Label(table, text=item, font=("Arial", 12), padx=10, pady=10, borderwidth=1, relief="solid")
                label.grid(row=i+1, column=j)

    def submit_attendance(self):
        # Get the attendance data from the input fields and tick marks
        faculty = self.faculty_entry.get()
        division = self.division_entry.get()
        date = self.date_entry.get()
        attendance_data = []
        for i, tick_var in enumerate(self.tick_buttons):
            roll = i+1
            name = self.student_labels[i].cget("text")
            attendance_status = "Present" if tick_var.get() else "Absent"
            attendance_data.append((faculty, division, date, name, attendance_status))
    
        # Insert the attendance data into the database
        sql = "INSERT INTO attendance (faculty, division, date, student_name, attendance_status) VALUES (%s, %s, %s, %s, %s)"
        mycursor.executemany(sql, attendance_data)
        mydb.commit()
    
        # Show a message in the GUI interface
        message_label = tk.Label(self.root, text="Attendance data inserted successfully", fg="green")
        message_label.grid(row=17, column=0, columnspan=3, pady=10) 

if __name__ == "__main__":
    app = AttendanceGUI()
    app.root.mainloop()
```

```python
import tkinter as tk

class AttendanceSystem:
    def __init__(self, master):
        self.master = master
        self.master.title("Attendance System")

        # Create labels and buttons
        self.label = tk.Label(self.master, text="Enter your name:")
        self.label.pack()
        self.entry = tk.Entry(self.master)
        self.entry.pack()
        self.button = tk.Button(self.master, text="Mark Attendance", command=self.mark_attendance)
        self.button.pack()

    def mark_attendance(self):
        name = self.entry.get()

        # Write name to a file
        with open("attendance.txt", "a") as file:
            file.write(name + "\n")

        # Display a message
        message = f"{name} is marked present."
        tk.messagebox.showinfo("Attendance", message)

# Create a Tkinter window
root = tk.Tk()

# Create the attendance system
attendance_system = AttendanceSystem(root)

# Start the main event loop
root.mainloop()

```

```python
import tkinter as tk
import mysql.connector as mysql
from tkinter import ttk
import tkinter.ttk as ttk

# Connect to MySQL database
db = mysql.connect(
    host="localhost",
    user="root",
    password="",
    database="rku_attendence_admin"
)

# Create a Tkinter window
window = tk.Tk()
window.title("RKU Attendance System")

# Create a label for the window
welcome_label = tk.Label(window, text="Welcome to RKU Server", font=("Arial", 18))
welcome_label.pack(pady=10)

# Create a label for the 'Faculty' input
faculty_label = tk.Label(window, text="Faculty", font=("Arial", 12))
faculty_label.pack(pady=5)

# Create an entry box for the 'Faculty' input
faculty_entry = tk.Entry(window, font=("Arial", 12))
faculty_entry.pack(pady=5)

# Create a label for the 'Date' input
date_label = tk.Label(window, text="Date", font=("Arial", 12))
date_label.pack(pady=5)

# Create an entry box for the 'Date' input
date_entry = tk.Entry(window, font=("Arial", 12))
date_entry.pack(pady=5)

# Create a label for the dropdown
division_label = tk.Label(window, text="Division", font=("Arial", 12))
division_label.pack(pady=5)

# Create a dropdown with two options: IT and CE
division_options = ["IT", "CE"]
division_var = tk.StringVar(window)
division_dropdown = tk.OptionMenu(window, division_var, *division_options)
division_dropdown.pack(pady=5)

 

def take_attendance():
    # Get the selected division
    division = division_var.get()

    # Get the students from the database
    cursor = db.cursor()
    cursor.execute(f"SELECT * FROM {division.lower()}_students")
    students = cursor.fetchall()

    # Create a new window for taking attendance
    attendance_window = tk.Toplevel(window)
    attendance_window.title(f"{division} Attendance")

    # Create a label for the window
    attendance_label = tk.Label(attendance_window, text=f"{division} Attendance", font=("Arial", 18))
    attendance_label.pack(pady=10)

    # Create a table for displaying attendance data
    attendance_table = ttk.Treeview(attendance_window, columns=("Roll No.", "Name", "Attendance"), show="headings")
    attendance_table.heading("Roll No.", text="Roll No.")
    attendance_table.heading("Name", text="Name")
    attendance_table.heading("Attendance", text="Attendance")
    attendance_table.pack(pady=10)

    # Add the student data to the table
    for student in students:
        attendance_table.insert("", "end", values=(student[0], student[1], ""))

        # Save the attendance in the database
        def save_attendance():
            attendance = 1 if attendance_var.get() else 0
            cursor.execute(f"INSERT INTO {division.lower()}_attendance (student_id, date, attendance) VALUES ({student[0]}, '{date_entry.get()}', {attendance})")
            db.commit()

            # Update the attendance data in the table
            attendance_table.item(student[0]-1, values=(student[0], student[1], "Present" if attendance_var.get() else "Absent"))

        # Create a checkbox for the student's attendance
        attendance_var = tk.BooleanVar()
        attendance_checkbox = tk.Checkbutton(attendance_window, variable=attendance_var)
        attendance_table.window_create(attendance_table.identify_cell(len(students)-1, 2), window=attendance_checkbox)

        # Create a button to save the attendance
        save_button = tk.Button(attendance_window, text="Save Attendance", command=save_attendance)
        attendance_table.window_create(attendance_table.identify_cell(len(students)-1, 3), window=save_button)
# Create a button to take attendance
attendance_button = tk.Button(window, text="Take Attendance", command=take_attendance)
attendance_button.pack(pady=10)

# Start the Tkinter event loop
window.mainloop()
```
