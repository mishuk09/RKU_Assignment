```python
import mysql.connector
from tkinter import *

# Connect to MySQL database
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="rku_attendence_admin"
)

# Create the main window
root = Tk()
root.title("RKU Student Attendance System")
root.geometry("500x400")

# Add a label with the university name
label = Label(root, text="Welcome to RKU Server", font=("Arial", 20))
label.pack(pady=20)

# Create a label for entering student name
label_name = Label(root, text="Enter student name:")
label_name.pack(pady=10)

# Create an entry field for student name
entry_name = Entry(root)
entry_name.pack()

# Create a label for entering student email
label_email = Label(root, text="Enter student email:")
label_email.pack(pady=10)

# Create an entry field for student email
entry_email = Entry(root)
entry_email.pack()

# Define a function to register a new student
def register_student():
    # Get the student name and email from the entry fields
    student_name = entry_name.get()
    student_email = entry_email.get()
    
    # Create a cursor to execute MySQL queries
    mycursor = mydb.cursor()
    
    # Execute a MySQL query to insert new student record
    sql = "INSERT INTO students (student_name, student_email) VALUES (%s, %s)"
    val = (student_name, student_email)
    mycursor.execute(sql, val)
    
    # Commit the changes to the database
    mydb.commit()
    
    # Clear the entry fields
    entry_name.delete(0, END)
    entry_email.delete(0, END)
    
    # Display a message box to confirm registration
    messagebox.showinfo("Registration", student_name + " registered successfully")

# Create a button for registering a new student
register_button = Button(root, text="Register New Student", command=register_student)
register_button.pack(pady=20)

# Run the main loop
root.mainloop()

```
